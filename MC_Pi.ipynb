# Estimating the value of Pi using Monte-Carlo method

#### Monte Carlo (MC) methods are a broad class of computational algorithms that rely on repeated random sampling to obtain numerical results. MC executes the same experiment (in our case, a computer model) with randomly selected inputs a large number of times, then extracts some summary statistic from the sum of outcomes.

#### Basic example of MC method is estimating the value of Pi.  

### Generate data points in random coordinate between x and y in [-1,1].

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#nPoints = 100
nPoints = 10000
#nPoints = 10000000
```

```python
circle = []
square = []

# Total Random numbers generated = possible x values* possible y values
for i in range(nPoints):

    # Randomly generated x and y values from a uniform distribution
    # Range of x and y values is -1 to 1
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)

    # Distance between (x, y) from the origin
    dist = x**2 + y**2

    # Checking if (x, y) lies inside the circle
    if dist <= 1:
      circle.append(np.array([x,y]))

    square.append(np.array([x,y]))

circle = np.array(circle)
square = np.array(square)
```

### Plot the points in circle

```python
plt.scatter(circle[:,0], circle[:,1], label='circle')
```

### Plot the points in the square

```python
plt.scatter(square[:,0], square[:,1], label='square')
```

### Plot the points in the circle and the square

```python
plt.scatter(square[:,0], square[:,1], label='square')
plt.scatter(circle[:,0], circle[:,1], label='circle')
plt.title('Number of datapoints = {}'.format(nPoints))
plt.legend(loc='upper left')
plt.show()
```

### Count the number of points in the circle and the square, and estimate pi

```python
nCircle = len(circle)
nSquare = len(square)

# Estimating value of pi,
# pi= 4*(no. of points generated inside the circle)/ (no. of points generated inside the square)
pi = 4 * nCircle / nSquare

print("Final Estimation of Pi= {:.2f}".format(pi))
```
